```python
import torch
import matplotlib.pyplot as plt
from sklearn.model_selection import validation_curve
%matplotlib inline
torch.set_printoptions(sci_mode=False,precision=4)
```

```python
with open('../data/amir_khan.txt', 'r') as file:
    names = file.read().splitlines()
```

```python
context_size = 256
embd_dim = 64
logging = False

device = torch.device("cpu" if torch.backends.mps.is_available() else "cpu")
batch_size = 64 if device.type == "mps" else 64
print(device, "| batch_size (before clamp):", batch_size)
```

```python
data = ''.join(names).lower()#[0:6000]

data = '_' * context_size + data
unique_text = sorted(set(data))
vocab_size = len(unique_text)

stoi = {text: index for index, text in enumerate(unique_text)}
itos = {index: text for index, text in enumerate(unique_text)}

encoder = lambda char_array: [stoi[char] for char in ''.join(char_array)]
decoder = lambda num_array: [itos[num] for num in num_array]
```

```python
print(stoi)
```

```python
x_numpy = []
y_numpy = []
temp_data = data

for t in range(len(temp_data) - context_size - 1):
    x_numpy.append(encoder(temp_data[t : t + context_size]))
    y_numpy.append(stoi[temp_data[t + context_size]])

x = torch.tensor(x_numpy, dtype=torch.long).to(device)
y = torch.tensor(y_numpy, dtype=torch.long).to(device)

N = x.shape[0]
train_ratio = 0.8
val_ratio = 0.1
test_ratio = 0.1

train_end = int(train_ratio * N)
val_end = int((train_ratio + val_ratio) * N)
perm = torch.randperm(N, device=device)
x = x[perm]
y = y[perm]

x_train, y_train = x[:train_end], y[:train_end]
x_val, y_val = x[train_end:val_end], y[train_end:val_end]
x_test, y_test = x[val_end:], y[val_end:]

x, y = x.to(device), y.to(device)

batch_size = min(batch_size, x.shape[0])
```

```python
from torch import nn
import torch.optim as optim

class Model(nn.Module):
    def __init__(self, embed_dim, hidden_layer_):
        super().__init__()
        self.embedding = nn.Embedding(vocab_size, embed_dim)

        self.linear_1 = nn.Linear(embd_dim * context_size, hidden_layer_, bias=False)
        self.bn1 = nn.BatchNorm1d(hidden_layer_)

        self.linear_2 = nn.Linear(hidden_layer_, hidden_layer_, bias=False)
        self.bn2 = nn.BatchNorm1d(hidden_layer_)

        self.linear_3 = nn.Linear(hidden_layer_, hidden_layer_, bias=False)
        self.bn3 = nn.BatchNorm1d(hidden_layer_)

        self.linear_4 = nn.Linear(hidden_layer_, vocab_size)
        self.activation = nn.Tanh()

    def forward(self, x_):
        emb = self.embedding(x_)
        emb = emb.view(emb.size(0), -1)

        l1 = self.linear_1(emb)
        l1 = self.bn1(l1)
        tanh = self.activation(l1)

        l2 = self.linear_2(tanh)
        l2 = self.bn2(l2)
        tanh_2 = self.activation(l2)

        l3 = self.linear_3(tanh_2)
        l3 = self.bn2(l3)
        tanh_3 = self.activation(l3)

        l4 = self.linear_4(tanh_3)
        return l4
```

```python
hidden_layers = [4, 8, 16, 20, 24]
hidden_layer = 8
loss = 0
criterion = nn.CrossEntropyLoss()
model = Model(embd_dim, hidden_layer).to(device)

optimizer = optim.Adam(model.parameters(), lr=0.001)
lr_iterations = []
lr_losses = []
iteration_count = 0
model.train()
# for epoch in range(100000):
#     idx = torch.randint(0, len(x_train), (batch_size,), device=device)
#     rand_x = x_train[idx]
#     rand_y = y_train[idx]
#
#     optimizer.zero_grad(set_to_none=True)
#     outputs = model(rand_x)
#     loss = criterion(outputs, rand_y)
#     loss.backward()
#     optimizer.step()
# print(hidden_layer, loss.item())
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

hidden_layers = [4, 8, 16, 32]

criterion = nn.CrossEntropyLoss()
results = []

for hidden_layer in hidden_layers:

    print(f"\nTraining with hidden_layer = {hidden_layer}")

    model = Model(embd_dim, hidden_layer).to(device)
    optimizer = optim.Adam(model.parameters(), lr=0.001)

    model.train()

    for epoch in range(100000):   # reduce for faster experimentation
        idx = torch.randint(0, len(x_train), (batch_size,), device=device)
        rand_x = x_train[idx]
        rand_y = y_train[idx]

        optimizer.zero_grad(set_to_none=True)
        outputs = model(rand_x)
        loss = criterion(outputs, rand_y)
        loss.backward()
        optimizer.step()

    model.eval()

    with torch.no_grad():
        train_loss = criterion(model(x_train), y_train).item()
        val_loss   = criterion(model(x_val), y_val).item()
        test_loss  = criterion(model(x_test), y_test).item()

    print(f"Hidden: {hidden_layer} | Train: {train_loss:.4f} | Val: {val_loss:.4f} | Test: {test_loss:.4f}")

    results.append({
        "hidden": hidden_layer,
        "train_loss": train_loss,
        "val_loss": val_loss,
        "test_loss": test_loss
    })
```
